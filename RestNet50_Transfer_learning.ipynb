#### Geting layers name of model

```python
import tensorflow as tf

#load the model
model=tf.keras.applications.ResNet50()

#Getting the name of all layers
layers=[layer.name for layer in model.layers]

#print the layers
print(layers)
print("Length:",len(layers))
```

#### Getting the numbers of parameters

```python
import tensorflow as tf

#load the model
model=tf.keras.applications.ResNet50()

#Getting the numbers of parameters
num_params=sum([tf.keras.backend.count_params(p) for p in model.trainable_weights])

#print the numbers of params
print("The numbers of parmas : ",num_params)
```

#### Getting class number

```python
import tensorflow as tf

#load the model
model=tf.keras.applications.ResNet50()

#Getting the numbers of class
num_classes=model.output_shape[-1]

#print the number of class
print("The numbers of class in the numbers:",num_classes)
```

#### Getting class name

```python
import tensorflow as tf

#load the model
model=tf.keras.applications.ResNet50()

#load imagenet class names
class_lables_path=tf.keras.utils.get_file('ImageNetLabels.txt',
                                         'https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')
with open(class_lables_path,"r") as f:
  class_names=f.read().splitlines()

#print the class name
print(class_names)
```

```python
image_array=image.img_to_array(img)
image_array.shape
```

```python
image_array_extend_dims=np.expand_dims(image_array,axis=0)
image_array_extend_dims.shape
```

```python
preprocessing_image=tf.keras.applications.inception_v3.preprocess_input(image_array_extend_dims)
preprocessing_image.shape
```
