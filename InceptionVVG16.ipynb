#### Getting layers name of the model

```python
#geting all the layers name
layers=[layer.name for layer in model.layers]

#print the layers name
print(layers)
print("layers: ", len(layers))
```

#### Getting the numbers of parameters

```python
#getting the numbers of parameters
num_params=sum([tf.keras.backend.count_params(n) for n in model.trainable_weights])

#print the numbers of params
print("The numbers of params : ", num_params)
```

#### Geting class number

```python
#Getting number of classes
num_classes=model.output_shape[-1]

#print output classes number
print("The number of output classes :",num_classes)
```

#### Getting class name

```python
#load imagenet class name
class_label_path=tf.keras.utils.get_file('ImageNetLabels.txt',
                                         'https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')

with open(class_label_path,"r") as f:
  class_names=f.read().splitlines()

#print class name
print(class_names)
```

```python
from keras.preprocessing import image
import numpy as np

#load image
img=image.load_img("/content/elephant.jpg", target_size=(224,224))

print("image size :",img.size)
print("image height: ", img.height)
print("image weight : ", img.width)
```

```python
image_array=image.img_to_array(img)
image_array.shape
```

```python
image_array_expend_dims=np.expand_dims(image_array, axis=0)
image_array_expend_dims.shape
```

```python
preprocessed_image=tf.keras.applications.vgg16.preprocess_input(image_array_expend_dims)
preprocessed_image.shape
```

```python
prediction_v2=model.predict(preprocessed_image)
prediction_v2.shape
```

```python
prediction_v2
```

```python
class_names[np.argmax(prediction_v2)+1]
```

```python
import keras

model=keras.applications.vgg16.VGG16()
prediction_v1=model.predict(preprocessed_image)
prediction_v1.shape
print(np.argmax(prediction_v1))
print(num_classes)
print(class_names[np.argmax(prediction_v1)+1])
```

#### Decode prediction

```python
from keras.applications import imagenet_utils
result=imagenet_utils.decode_predictions(prediction_v1)
result
```
